<h3>Plotting with matplotlib</h3>
<p>In Python <b>matplotlib</b> is the most important package for plotting. Here, you will just get started with a few plots in <i>matplotlib</i> but you can have a look in the <a href="https://matplotlib.org/stable/gallery/index.html" target="_blank">matplotlib gallery</a> to get a sense of what you can do.</p>
<p>First you will get a few examples:</p>


Plot the function $f(x) = x^2$ on the interval [-4 5] (run the code in the cell):

```python
x = np.arange(-4,5.1,1) # Create x-axis (-4 -3 ... 4 5)
fx = x**2               # Compute the function x^2 in the x-values
# Plot the function
plt.plot(x,fx)
plt.title('f(x) = x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()    # tells Python we are ready to show the plot
```

<p>If you look at the graph, it's a bit jagged. The reason is that the function values are plotted only in the $x$-values (with straight lines in between), and the the step between the numbers is too high. I cases like this, it makes sense to use <code>np.linspace</code>, as we don't have a specified increment. Copy the code, paste it in the  cell below and replace <code>np.arange</code> with <code>np.linspace</code>. Use e.g. 200 points (elements) in $x$.
    </p>

```python
# Enter your code here
x = np.linspace(-4,5.1,200) # Create x-axis (-4 -3 ... 4 5)
fx = x**2               # Compute the function x^2 in the x-values
# Plot the function
plt.plot(x,fx)
plt.title('f(x) = x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()    # tells Python we are ready to show the plot
```

<p>If you've done it correctly, the graph should be more smooth compared with the first plot.</p>
<p>
Plot the function $f(x)=x^2 \cdot cos(x)$ on the interval [-3 3]
</p>

```python
x = np.linspace(-3,3,200) # Create x-axis, -3 to 3 in 200 points
fx = (x**2)*np.cos(x)   # Compute the function x^2*cos(x) in the x-values
# Plot the function
plt.plot(x,fx)
plt.title('f(x) = x^2 * cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

<p>One thing to think about here: the function <code>(x**2)*np.cos(x)</code> is actually multiplication between two vectors. The first vector is <code>x**2</code> and the 2nd is <code>np.cos(x)</code>. In part 3 (next part) of this lab we will see that matrix/vector multiplication can be performed in two different ways. It's either a dot-product, or multiplication element by element. As we for every <tt>x</tt>-value are supposed to calculate the formula, it is in this case elementwise multiplication. The use of <tt>*</tt> rather than <tt>@</tt> in the multiplication leads to elementwise multiplication.<br>
What we do here is actually equivalent to a for-loop of the type
    <pre>
    for i in range(len(x)):
      fx[i] = (x[i]**2)*np.cos(x[i])
    </pre>
</p>
<p>
The use of elementwise multiplication is replacing a for-loop (and is also more efficient than using a for-loop).
</p>

<p>Plot the two functions $f(x)=x^2$ and $cos(x)$ on the interval [-3 3], and in the same plot
    </p>

```python
x = np.linspace(-3,3,200) # Create x-axis, -3 to 3 in 200 points
fx1 = x**2
fx2 = np.cos(x)
# Plot the functions
plt.plot(x,fx1)
plt.plot(x,fx2)
plt.title('f(x) = x^2 and cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['x^2','cos(x)'])
plt.show()
```

<p>
    When we have several graphs in the same plot, use <tt>plt.legend</tt> so we know which graph is which. The syntax is <tt>plt.legend([first plot, 2nd plot])</tt>, i.e the same order as in the <tt>plt.plot</tt>-commands.
    </p>
    <p>
  You can change markers and color on the markers, by adding it in the plot-command. For example <code>plt.plot(x,fx1,'g--')</code> stands for green (<tt>g</tt>) and dashed (<tt>--</tt>), and <code>plt.plot(x,fx2,'y:')</code> stands for yellow (<tt>y</tt>) and dotted (<tt>:</tt>). <br>
    Try that change in the box below (copy the code, paste it in the cell and do the changes).
    </p>

```python
# Enter your code here
x = np.linspace(-3,3,200) # Create x-axis, -3 to 3 in 200 points
fx1 = x**2
fx2 = np.cos(x)
# Plot the functions
plt.plot(x,fx1,'g--')
plt.plot(x,fx2,'y:')
plt.title('f(x) = x^2 and cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['x^2','cos(x)'])
plt.show()
```
